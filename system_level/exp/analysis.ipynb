```python
cpu = [0.09, 0.29, 0.31, 0.97, 4.96, 68.5, 848.22, 1833.19]
gpu = [0.00038, 0.0013, 0.0023, 0.0031, 0.018, 0.17, 1.03, 2.11]
ratio = [cpu[i]/gpu[i] for i in range(len(cpu))]

# plot the ratio
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(8,5), dpi=200)
plt.plot(ratio, 'o-')
plt.xlabel('matrix size')
plt.ylabel('speedup')
plt.title('Speedup of GPU over CPU')
plt.show()
```
