```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# only plot from epoch 1 on (leave epoch 0 alone)

with open('./medoflow_results.json') as f:
    medoflow_results = json.load(f)
    
with open('./pytorch_results.json') as f:
    pytorch_results = json.load(f)
    
with open('./tensorflow_results.json') as f:
    tensorflow_results = json.load(f)
```

```python
medoflow_results['train_acc'] = medoflow_results['train_acc'][1:]
medoflow_results['val_acc'] = medoflow_results['val_acc'][1:]
medoflow_results['loss'] = medoflow_results['loss'][1:]

pytorch_results['train_acc'] = pytorch_results['train_acc'][1:]
pytorch_results['val_acc'] = pytorch_results['val_acc'][1:]
pytorch_results['loss'] = pytorch_results['loss'][1:]

tensorflow_results['train_acc'] = tensorflow_results['train_acc'][1:]
tensorflow_results['valid_acc'] = tensorflow_results['valid_acc'][1:]
tensorflow_results['loss'] = tensorflow_results['loss'][1:]
```

```python
fig = plt.figure(figsize=(15, 5), dpi=200)
axs = fig.subplots(1, 3,)

x = np.arange(1, 20)

axs[0].plot(x, medoflow_results['train_acc'], label='MedoFlow')
axs[0].plot(x, pytorch_results['train_acc'], label='PyTorch')
axs[0].plot(x, tensorflow_results['train_acc'], label='TensorFlow')
axs[0].set_title('Training Accuracy')
axs[0].set_xlabel('Epoch')
# x interval should be 1
axs[0].set_xlim(0, 21)
axs[0].set_xticks(np.arange(0, 21, 5))
axs[0].set_ylabel('Accuracy')
axs[0].legend()

axs[1].plot(x, medoflow_results['val_acc'], label='MedoFlow')
axs[1].plot(x, pytorch_results['val_acc'], label='PyTorch')
axs[1].plot(x, tensorflow_results['valid_acc'], label='TensorFlow')
axs[1].set_title('Validation Accuracy')
axs[1].set_xlabel('Epoch')
axs[1].set_xlim(0, 21)
axs[1].set_xticks(np.arange(0, 21, 5))
axs[1].set_ylabel('Accuracy')
axs[1].legend()

axs[2].plot(x, medoflow_results['loss'], label='MedoFlow')
axs[2].plot(x, pytorch_results['loss'], label='PyTorch')
axs[2].plot(x, tensorflow_results['loss'], label='TensorFlow')
axs[2].set_title('Loss')
axs[2].set_xlabel('Epoch')
axs[2].set_xlim(0, 21)
axs[2].set_xticks(np.arange(0, 21, 5))
axs[2].set_ylabel('Loss')
axs[2].legend()
 
```
